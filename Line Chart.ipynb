## ***Cénario:***
### **Imagine que você gerência uma equipe de tecnologia da informação (TI). Sua equipe recebe tíquetes, ou questões técnicas, dos funcionários. No ano passado, duas pessoas saíram da equipe e, na época, você decidiu não substitui-las. Você tem ouvido rumorosas reclamações dos demais funcionários sobre terem de "preencher a lacuna". Recentemente, você foi questionado sobre suas necessidades de contratação para o próximo ano e está se perguntando se deve contratar mais duas pessoas. Você coloca em um gráfico a tendência mensal dos tíquetes recebidos e aqueles processados no ano civil passado. Você vê que há certa evidência de que a produtividade de sua equipe está sendo prejudicada pela falta de pessoal e agora quer tomar o visual que criou a base para seu pedido de contratação.**

### **Dado o Cénario, vamos para as Análises e Melhores Métodos de Visualização**

## **Iniciando**
### **Importando Bibliotecas para Dados e Visualização**

```python
import matplotlib.pyplot as plt
```

## **Definindo os Dados**
### **- Meses do Ano**
### **- Volume de Tíquetes Recebidos**
### **- Volume de Tíquetes Processados**

```python
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
tiquetes_recebidos = [152, 173, 289, 149, 167, 155, 140, 200, 153, 147, 150, 172]
tiquetes_processados = [152, 173, 288, 149, 167, 149, 134, 151, 136, 107, 129, 148]
```

## **Criando Gráfico de Linha para Tíquetes Recebidos e Processados ao Longo do Ano**

```python
plt.figure(figsize=(10,7))
plt.plot(meses, tiquetes_processados, color='forestgreen', marker='o', label='Processados')
plt.plot(meses, tiquetes_recebidos, color='springgreen',marker='x', label='Recebidos')
plt.legend(loc='lower center')
plt.xlabel("Meses")
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

### **Ao executar o gráfico, percebe algo em relação a Visualização e Storytelling com Dados? Veja a Saturação, Organização, Contexto, Atratividade, Clareza, Contrastes, etc. Processados? Recebidos? (???). Percebe que não é agradavel visualizar o gráfico em relação a esses parâmetros?**
### **Isso porque as técnicas não foram aplicadas juntos aos métodos de orientação da visualização, com isso, vamos ajustar o gráfico para que fique legivel, atrativo, organizado e de acordo com as melhores práticas para Visualização de Dados.**
#### ***Spoiler: Faremos uma Versão Dark Theme do Gráfico para uma Diversificação de Estilo de Layout***

## **1 - Ajuste as Cores do Gráfico**
### **Para uma Visualização Eficaz, as cores do gráfico é de extrema importância, pois, as cores obtém carga cognitiva no cérebro humando aponto de observarmos e nós mesmos decidirmos se seguiremos a visualizar e analisar o gráfico, ou apenas ignorar aquele monte de informações não organizadas e confusas.**

### **Dado o Primero Método, vamos ajustar as cores e definifir a primeira melhora da visualização**

```python
plt.figure(figsize=(10,7))
plt.plot(meses, tiquetes_processados, color='tomato', marker='o', label='Processados')
plt.plot(meses, tiquetes_recebidos, color='royalblue',marker='x', label='Recebidos')
plt.legend(loc='lower center')
plt.xlabel("Meses")
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

### **Percebe agora como são importantes as cores para a Visualização de Dados? Veja como Evoluiu a Visualização do Gráfico e Legibilidade apenas ajustando as Cores.**
### **Uma Dica para Cores, se você deseja destacar uma informação não use cores super saturadas e fortes demais, isso torna a visualização desagradavél e cansativa, em vez disso, use tons pasteis, de preferência mais claros e de acordo com o que pretende destacar.**
### **Isso torna o gráfico mais legivel, mais atratativo, mais organizado e mais coerente com as informações descritas. Mas calma, ainda não terminamos, falta alguns processos extremante relevantes para o Storytelling**

## **2 - Títulos, Legendas & Marcadores**
### **Percebeu que nos gráficos anteriores não há uma coerência da informação a ser transmitida? Percebe que os detalhes realmente faz a diferença? Análise novamente, e perceba as estruturações nada harmoniozas.**
### **Nesta segunda etapa iremos dar enfaze nos detalhes e melhorar drasticamente a visualização.**
### **Upgrade no Gráfico - Título, Legendas e Marcadores**

```python
plt.figure(figsize=(10,7))

plt.title("Tíquetes Processados e Recebidos ao Longo do Ano\n", fontfamily='sans-serif', fontsize=15, fontweight='bold')
plt.plot(meses, tiquetes_processados, color='tomato', label='Processados', linewidth=2)
plt.plot(meses, tiquetes_recebidos, color='royalblue', label='Recebidos', linewidth=2)

plt.xlabel("Meses", fontsize=14, fontweight='bold', fontfamily='serif')
plt.xticks(rotation=45, fontweight='bold', family='serif')
plt.yticks(fontweight='bold', family='sans-serif')
plt.grid(axis='y', linewidth=0.4)
plt.legend(loc='best', fontsize=11)

plt.show()
```

#### **Veja agora a Diferença Gritante de um Gráfico para o outro. As Letras estão mais Estruturadas, os Marcadores das Linhas dava uma perspectiva mais rude e desnecessária então eliminamos, a legenda do gráfico esta precisa e hármonioza com o espaço do gráfico, tornando-o coerente e agradavel a visualização, e o título nos da uma base e noção do que se trata as informações do Gráfico de Linha viabilizando o entendimento e direcionamento do que analisarmos diente as linhas e os meses do ano.**
#### **Observe a melhora essencial da Visualização de Dados. Através dos detalhes é que transmitimos a informação de forma legivel e estruturada, sem a Organização Gráfica, Contexto e Clareza é impossivel passarmos uma boa Informação através dos Dados e da Visualização.**
#### **Vamos seguir para o próximo método, e esse será breve porém de extrema importância.**

## **3 - Rotação de Eixo & Linhas de Grade**
### **Agora nós vamos eliminarmos a carga cognitiva extrínseca. Veja que o último gráfico esta com todos os meses escritos sem abreviação e todos na diagonal de 45 graus, e também abaixo dos meses esta escrito "Meses" no eixo x, dando um indicativo obvio e irrelevante, isso também é desnecessário para a visualização, até porque nos tendemos a lermos o eixo x primeiro do que legenda abaixo dele, o que já saberiamos que se trata dos meses ao lermos. Enquanto as Linhas de Grade elas são optativas, mas cuidado, você nunca deve destacar a ponto de estar mais visivel e chamativo que os Dados e a Informação que deseja passar. São subparâmetros que você deve ter cuidado ao usar, se destarcar demais, provavelmente, irá perder o contexto, clareza e organização, e com isso você perdeu a Visualização de Dados.**
### **Portanto, vamos eliminar o Indicativo "Meses" por ser óbvio e desnecessário, e também vamos abreviar todos os meses e coloca-los em uma posição de melhor legibilidade, além disso, vamos decartar as Linhas de Grade para clareza e suavidade na transmissão de informação.**

```python
meses_novo = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
```

```python
plt.figure(figsize=(10,7))

plt.title("Tíquetes Processados e Recebidos ao Longo do Ano\n", fontfamily='sans-serif', fontsize=15, fontweight='bold')
plt.plot(meses_novo, tiquetes_processados, color='tomato', label='Processados', linewidth=2.5)
plt.plot(meses_novo, tiquetes_recebidos, color='royalblue', label='Recebidos', linewidth=2.5)

plt.xticks(fontsize=12, fontweight='bold', family='serif')
plt.yticks(fontweight='bold', family='sans-serif')
plt.legend(loc='best', fontsize=11)

plt.show()
```

### **Veja a evolução do visual, veja os traços melhorados junto com os ângulos.**
### **Geralmente, elementos diagonais, como linhas e texto, devem ser evitados. Eles parecem desordenados e, no caso do texto, são mais difíceis de ler que seus equivalentes horizontais. No que diz respeito a orientação de texto, um estudo (Wigdor e Balakrishnan, 2005) descobriu que a leitura de texto inclinado a 45 graus em qualquer direção era, em média, 52% mais lenta que a leitura de texto com orientação normal (para texto inclinado a 90 graus em qualquer direção era 205% mais lenta, em média). É melhor evitar elementos diagonais no layout.**

## **4 - Linha de Base Igual a Zero**
### **Devido ao modo como nossos olhos comparam pontos extremos relativos no gráfico de linha, é importante que a linha de base esteja igual a zero (em que o eixo x cruza o eixo y em zero), caso contrário você obterá uma comparação visual falsa.**
### **É primordial ter o contexto da linha inteira para se fazer uma comparação precisa e correta.**

```python
plt.figure(figsize=(9,6))

plt.title("Tíquetes Processados e Recebidos ao Longo do Ano\n", fontfamily='sans-serif', fontsize=15, fontweight='bold')
plt.plot(meses_novo, tiquetes_processados, color='tomato', label='Processados', linewidth=2.5)
plt.plot(meses_novo, tiquetes_recebidos, color='royalblue', label='Recebidos', linewidth=2.5)

# Configurar o limite inferior do eixo y para começar em zero
plt.ylim(0, max(max(tiquetes_recebidos), max(tiquetes_processados)) + 10)

plt.xticks(fontsize=12, fontweight='bold', family='serif')
plt.yticks(fontweight='bold', family='sans-serif')
plt.legend(loc='best', fontsize=11)

plt.tight_layout()
plt.show()
```

### **Observe como melhorou a amplitude e a comparação contextual referênte no gráfico.**

## **5 - Borda & Legenda Direta**

### **Bordas de gráfico são normalmente desnecessárias, em vez disso vamos usar o espaço em branco para diferenciar o visual de outros elementos na página, conforme necessário.**
### **Agora que eliminamos grande parte da carga cognitiva extrínseca, o trabalho de ir e voltar entre a legenda e os dados é ainda mais evidente. Queremos identificar tudo que pareça esforço para nosso público e assumir nossa responsabilidade como designers da informação. Neste caso, podemos usar o princípio da proximidade e colocar as legendas bem perto dos dados que elas descrevem.**

```python
plt.figure(figsize=(9,6))

plt.title("Tíquetes Processados e Recebidos ao Longo do Ano\n\n", fontfamily='sans-serif', fontsize=15, fontweight='bold')
plt.plot(meses_novo, tiquetes_processados, color='tomato', label='Processados', linewidth=2.5)
plt.plot(meses_novo, tiquetes_recebidos, color='royalblue', label='Recebidos', linewidth=2.5)

# Configurar o limite inferior do eixo y para começar em zero
plt.ylim(0, max(max(tiquetes_recebidos), max(tiquetes_processados)) + 10)

# Adicionar legendas no final das linhas
for i, txt in enumerate(['Processados', 'Recebidos']):
    y = tiquetes_processados[-1] if txt == 'Processados' else tiquetes_recebidos[-1]
    plt.text(meses_novo[-1], y, f'  {txt}', fontsize=11, fontweight='bold', color='tomato' if txt == 'Processados' else 'royalblue', verticalalignment='center')
    
# Remover as bordas superior e direita
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xticks(fontsize=12, fontweight='bold', family='serif')
plt.yticks(fontweight='bold', family='sans-serif')

plt.tight_layout()
plt.show()
```

### **Agora veja um layout digno de informação e visualização eficaz. Eliminamos tudo de poluído e cargas densas de cognição extrínseca, tudo isso torna a legibilidade, o contexto, a harmonia da informação e saturação totalmente adequada para a visualização de dados e coerência de informação.**
